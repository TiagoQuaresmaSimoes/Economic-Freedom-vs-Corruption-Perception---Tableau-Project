```python
import pandas as pd
```

```python
eco_free = pd.read_excel(r'C:\Users\Quaresma\Documents\IRONHACK\Tableau project\Datasets\Economic_Freedom_index_data.xlsx')
cpi= pd.read_excel(r'C:\Users\Quaresma\Documents\IRONHACK\Tableau project\Datasets\Corruption_Perception_index.xlsx')
```

```python
def merge_and_clean_data(df1, df2):
    def norm_headers(data):
        col = [data.columns[i].lower().replace(" ", "_") for i in range(len(data.columns))]
        data.columns = col

    norm_headers(df1)
    norm_headers(df2)
    df_all = pd.merge(df1, df2, how='inner', on=['code', 'year'])
    
    df_all = df_all.drop(['countryid', 'country_name', 'region_y'], axis=1)
    df_all.rename(columns={
        "region_x": "region",
        "rank": "cpi_rank",
        "standard_error": "cpi_standard_error",
        "world_rank": "economic_freedom_world_rank",
        "region_rank": "economic_freedom_region_rank",
        'score': 'eco_free_score'
    }, inplace=True)

    return df_all
```

```python
df_all=merge_and_clean_data(cpi, eco_free)
```
